```python
import numpy as np
import matplotlib.pyplot as plt
import os
import fitpatterns_opt_drift as fitpatterns
```

#### Set up behavioral patterns and trial structure for K02

```python
# load K02 dataset exp 1
ll = 10
lag_examine = 4
LSA_path = 'data/K02_files/K02_LSA.txt'
data_path = 'data/K02_files/K02_data.txt'
data_rec_path = 'data/K02_files/K02_recs.txt'
subjects_path = 'data/K02_files/K02_subject_ids.txt'

data_pres = np.loadtxt(data_path, delimiter=',')
data_rec = np.loadtxt(data_rec_path, delimiter=',')
  
#extract behavioral patterns    
_,data_recalls,data_sp = fitpatterns.data_recode(data_pres, data_rec)
data_spc,data_pfr = fitpatterns.get_spc_pfr(data_sp,ll)
data_crp = fitpatterns.get_crp(data_sp,lag_examine,ll)
np.savetxt('data/K02_files/K02_spc.txt',data_spc)
np.savetxt('data/K02_files/K02_pfr.txt',data_pfr)
np.savetxt('data/K02_files/K02_crp.txt',data_crp)


#data_spc = np.loadtxt('data/K02_files/K02_spc.txt')
#data_pfr = np.loadtxt('data/K02_files/K02_pfr.txt')
#data_crp = np.loadtxt('data/K02_files/K02_crp.txt')

# set all lists to be independent
subjects = np.loadtxt(subjects_path, delimiter=',')
subjects_new = [i for i,x in enumerate(subjects)]
np.savetxt('data/K02_files/K02_list_ids.txt',subjects_new)
```

```python
LL = 10
percent = 50
accs = []
subjects_unique = np.unique(subjects)
for s, subj in enumerate(subjects_unique):
    recall_sp = [x for i,x in enumerate(data_sp) if subjects[i]==subj] # all trials
    acc = np.mean([len(np.unique([x for x in rp if 0<=x<LL])) for rp in recall_sp]) 
    accs.append(acc)
threshold = np.percentile(accs,percent)

data_sp_good = []
data_sp_bad = []
subjects_good = []
subjects_bad = []
for s, subj in enumerate(subjects_unique):
    recall_sp = [x for i,x in enumerate(data_sp) if subjects[i]==subj] # all trials
    acc = np.mean([len(np.unique([x for x in rp if 0<=x<LL])) for rp in recall_sp]) 
    if acc>threshold:
        data_sp_good = data_sp_good + recall_sp
        
    else:    
        data_sp_bad = data_sp_bad + recall_sp   
        
    
spc_g, pfr_g = fitpatterns.get_spc_pfr(data_sp_good,LL)
crp_g = fitpatterns.get_crp(data_sp_good,4,LL)
spc_b, pfr_b = fitpatterns.get_spc_pfr(data_sp_bad,LL)
crp_b = fitpatterns.get_crp(data_sp_bad,4,LL)

np.savetxt('data/K02_files/K02_spc_good.txt',spc_g)
np.savetxt('data/K02_files/K02_pfr_good.txt',pfr_g)
np.savetxt('data/K02_files/K02_crp_good.txt',crp_g)
subjects_good = [i for i,x in enumerate(data_sp_good)]
np.savetxt('data/K02_files/K02_list_ids_good.txt',subjects_good)

np.savetxt('data/K02_files/K02_spc_bad.txt',spc_b)
np.savetxt('data/K02_files/K02_pfr_bad.txt',pfr_b)
np.savetxt('data/K02_files/K02_crp_bad.txt',crp_b)
subjects_bad = [i for i,x in enumerate(data_sp_bad)]
np.savetxt('data/K02_files/K02_list_ids_bad.txt',subjects_bad)

plt.rcParams['figure.figsize'] = (16,5.5)
plt.subplot(1,3,1)
plt.plot(spc_g)
plt.plot(spc_b)
plt.subplot(1,3,2)
plt.plot(pfr_g)
plt.plot(pfr_b)
plt.subplot(1,3,3)
plt.plot(crp_g)
plt.plot(crp_b)
plt.legend(['Top half','Bottom half'])
plt.show()    
```
