### optimal-CMR

```python
# data recoding 
def data_recode(data_pres,data_rec):
    # recode data into lists without 0-paddings; convert to python indexing from matlab indexing by -1
    presents = [[int(y)-1 for y in list(x) if y>0] for x in list(data_pres)]
    recalls = [[int(y)-1 for y in list(x) if y>0] for x in list(data_rec)]

    # recode recall data into serial positions (set as x+100 if it is a repetition; set as -99 if it is an intrusion)
    recalls_sp = []
    for i,recall in enumerate(recalls):
        recall_sp = []
        this_list = presents[i]
        recallornot = np.zeros(len(this_list))
        for j,item in enumerate(recall):
            try:
                sp = this_list.index(item)
                if recallornot[sp]==0: # non-repetition
                    recall_sp.append(this_list.index(item))
                    recallornot[sp]=1
                else: # repetition
                    recall_sp.append(this_list.index(item)+100)
            except:   # intrusion 
                recall_sp.append(-99)

        recalls_sp.append(recall_sp)    
    return presents,recalls,recalls_sp


def obj_func_subj(beta_enc,beta_rec,gamma_fc,phi_s,phi_d,epsilon_d,k,primacy,ll):  
    """Error function that we want to minimize"""
    ###############
    #
    #   Load parameters and path
    #
    ###############
    #LSA_path = '/mnt/bucket/people/qiongz/optimalmemory/pyCMR2/K02_files/K02_LSA.txt'
    data_path = 'data/K02_temp_data_structure.txt'
    data_rec_path = 'data/K02_temp_data_structure.txt'
    subjects_path = 'data/K02_temp_list_ids.txt' # assume each list is a subject
    LSA_mat = np.zeros([ll,ll])
    for i in range(ll-1):
        LSA_mat[i+1][i]=1
    #LSA_mat, data_path, data_rec_path, subjects_path = load_structure(-1) # where data structure is from

    # current set up is for fitting the non-emot version of the model
    param_dict = {  
        'beta_enc': 1,#beta_enc,           # rate of context drift during encoding
        'beta_rec': 1,#beta_rec,           # rate of context drift during recall
        'beta_rec_post': 1.0,      # rate of context drift between lists
                                        # (i.e., post-recall)

        'gamma_fc': 0.1,#gamma_fc,  # learning rate, feature-to-context
        'gamma_cf': 0.5,  # learning rate, context-to-feature
        'scale_fc': 1 - gamma_fc,
        'scale_cf': 0.0,

        's_cf': 2,       # scales influence of semantic similarity
                                # on M_CF matrix

        's_fc': 0,            # scales influence of semantic similarity
                                # on M_FC matrix.
                                # s_fc is first implemented in
                                # Healey et al. 2016;
                                # set to 0.0 for prior papers.

        'phi_s': 0,#phi_s,      # primacy parameter
        'phi_d': phi_d,      # primacy parameter

        'epsilon_s': 0.0,      # baseline activiation for stopping probability 
        'epsilon_d': epsilon_d,        # scale parameter for stopping probability 
        

        'k': k+5,   # luce choice rule scale 
        
        'cue_position':-1, # no initial cue
        'primacy':primacy, # control whether to start from the beginning (1) or the end (0)
        'enc_rate':1, # encoding rate
    }
    # run probCMR on the data
    resp, times,_ = CMR_prob.run_CMR2(
            recall_mode=0, LSA_mat=LSA_mat, data_path=data_path,rec_path = data_rec_path,
            params=param_dict, subj_id_path=subjects_path, sep_files=False)
    data_pres = np.loadtxt(data_path, delimiter=',')
    _,CMR_recalls,CMR_sp = data_recode(data_pres, resp)
    acc = np.mean([len(np.nonzero(x)[0]) for x in resp])
    return CMR_sp,acc
```

```python
N = 2 # Trials of 10 subjects
ll = 10
itr = 12
num = 200
primacy = 0

# load CMR parameters
optimizer = load_model.load_vanilla()  
load_logs(optimizer, logs=["./output/logs1001_fitCMRprob_K02Behav_poi02.json"])
param = optimizer.max['params']

firstscore = []
firstscore2 = []
scores = np.zeros([num,itr+1])
scores2 = np.zeros([num,itr+1])
accs = np.zeros([num,itr])
data1 = []
data2 = []
for n in range(num):
    data = []
    # set up the initial random trial
    for i in range(N):
        a = [int(j+1) for j in range(ll)]
        random.shuffle(a)
        data.append(a)
    np.savetxt('data/K02_temp_data_structure.txt',data,delimiter=',')
    lists = [int(i+1) for i in range(N)]
    np.savetxt('data/K02_temp_list_ids.txt',lists)
    #print(data)

    count = 0
    count2 = 0
    for j in range(len(a)-1):     
        if a[j] +1 == a[j+1]:
            count = count + 1              
        count2 = count2 + np.abs(a[j]-a[j+1])
    scores[n][0] = count        
    scores2[n][0] = count2
    
    if n<2:
        print(data)
    for i in range(itr):

        # recall given study list
        #print(np.min([primacy+0.2*count,1]))
        CMR_sp,acc  = obj_func_subj(param['beta_enc'],param['beta_rec'],param['gamma_fc'],param['phi_s'],param['phi_d'],param['epsilon_d'],param['k'],np.min([primacy+0.2*count,1]),ll+1)
        #CMR_sp,acc  = obj_func_subj(param['beta_enc'],param['beta_rec'],param['gamma_fc'],param['phi_s'],param['phi_d'],param['epsilon_d'],param['k'],primacy,ll+1)
        accs[n][i] = acc
        if i ==0:
            data1.append(CMR_sp[0])
        else:
            data2.append(CMR_sp[0])
        count = 0
        count2 = 0
        seq = [int(j+1) for j in range(ll)]
        seq = set(seq)
        recall = []
        #print('data')
        #print(data[0])
        #print('recalled')
        #print(CMR_sp[0])
        #print(' ')
        for j,x in enumerate(CMR_sp[0]):
            seq.remove(data[0][x])
            #print(seq)
            recall.append(data[0][x])
        temp = list(seq)
        random.shuffle(temp)
        #temp.extend(recall)
        recall.extend(temp)
        temp = recall.copy()
        
        if n<2:
            print(i)
            print(CMR_sp[0])
            print(temp)
        
        for j in range(len(temp)-1):     
            if temp[j] +1 == temp[j+1]:
                count = count + 1              
            count2 = count2 + np.abs(temp[j]-temp[j+1])
        scores[n][i+1] = count        
        scores2[n][i+1] = count2


        # study material for next iteration
        data = [[x for x in temp] for k in range(N)]
        np.savetxt('data/K02_temp_data_structure.txt',data,delimiter=',')
        lists = [int(i+1) for k in range(N)]
        np.savetxt('data/K02_temp_list_ids.txt',lists)
        #print(data)
```
