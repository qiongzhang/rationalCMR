```python
# import package
from bayes_opt import BayesianOptimization
from bayes_opt.observer import JSONLogger # for saving 
from bayes_opt.event import Events # for saving
from bayes_opt.util import load_logs # for loading


# project specific
import numpy as np
import matplotlib.pyplot as plt
import os
#import fitpatterns_autolab_simple as fitpatterns
import fitpatterns_opt_drift as fitpatterns
import itertools
from load_module_prob import load_module
```

```python
def compareModelData(model,optimizer): # 0-1-2-3-4-9-7
    param = optimizer.max['params']
    if model==0: # full model 
        RMSE = fitpatterns.obj_func_simple_model0(param['beta_enc'],param['beta_rec'],param['beta_rec_post'],param['gamma_fc'],param['gamma_cf'],param['s_cf'],param['s_fc'],param['phi_s'],param['phi_d'],param['epsilon_s'],param['epsilon_d'],param['k_intra'])
    if model==1: # simplified stopping probability (2->1, epsilon_s = 0)
        RMSE = fitpatterns.obj_func_simple_model1(param['beta_enc'],param['beta_rec'],param['beta_rec_post'],param['gamma_fc'],param['gamma_cf'],param['s_cf'],param['s_fc'],param['phi_s'],param['phi_d'],param['epsilon_d'],param['k_intra'])
    if model==2: # simplified semantic effects during encoding (s_fc = 0)
        RMSE = fitpatterns.obj_func_simple_model2(param['beta_enc'],param['beta_rec'],param['beta_rec_post'],param['gamma_fc'],param['s_fc'],param['phi_s'],param['phi_d'],param['epsilon_s'],param['epsilon_d'],param['k_intra'])
    if model==3: # simplified semantic effects during recall (gamma_cf = 0, s_cf = 0)
        RMSE = fitpatterns.obj_func_simple_model3(param['beta_enc'],param['beta_rec'],param['beta_rec_post'],param['gamma_fc'],param['gamma_cf'],param['s_cf'],param['phi_s'],param['phi_d'],param['epsilon_s'],param['epsilon_d'],param['k_intra'])
    if model==4: # simplified primacy (2->1, phi_d = 10)
        RMSE = fitpatterns.obj_func_simple_model4(param['beta_enc'],param['beta_rec'],param['beta_rec_post'],param['gamma_fc'],param['gamma_cf'],param['s_cf'],param['s_fc'],param['phi_s'],param['epsilon_s'],param['epsilon_d'],param['k_intra'])
    if model==5: # m1 + m2 + m3
        RMSE = fitpatterns.obj_func_simple_model5(param['beta_enc'],param['beta_rec'],param['beta_rec_post'],param['gamma_fc'],param['phi_s'],param['phi_d'],param['epsilon_d'],param['k_intra'])
    if model==6: # m3 + beta_rec=1 + m9
        RMSE = fitpatterns.obj_func_simple_model6(param['beta_enc'],param['gamma_fc'],param['gamma_cf'],param['s_cf'],param['phi_s'],param['phi_d'],param['epsilon_s'],param['epsilon_d'],param['k_intra'])
    if model==7:  #m3 + m1 + m9
        RMSE = fitpatterns.obj_func_simple_model7(param['beta_enc'],param['beta_rec'],param['gamma_fc'],param['gamma_cf'],param['s_cf'],param['phi_s'],param['phi_d'],param['epsilon_d'],param['k_intra'])
    if model==8: 
        RMSE = fitpatterns.obj_func_simple_model8(param['beta_enc'],param['beta_rec'],param['gamma_fc'],param['gamma_cf'],param['b_cf'],param['s_cf'],param['phi_s'],param['phi_d'],param['epsilon_d'],param['k_intra'])
    if model==9: # beta_post = 1
        RMSE = fitpatterns.obj_func_simple_model9(param['beta_enc'],param['beta_rec'],param['gamma_fc'],param['gamma_cf'],param['s_cf'],param['s_fc'],param['phi_s'],param['phi_d'],param['epsilon_s'],param['epsilon_d'],param['k_intra'])
    return RMSE
```

```python
model = 7
optimizer = load_module(0)
load_logs(optimizer, logs=["./output/logs0826_fitCMRprob2modes_Z19_ucb5.json"])
print(optimizer)
print(optimizer.max)

plt.rcParams['figure.figsize'] = (10,4)
y_obs = np.array([res["target"] for res in optimizer.res])
plt.plot(y_obs)
plt.show()

plt.rcParams['figure.figsize'] = (20,10)
print("New optimizer is now aware of {} points.".format(len(optimizer.space)))
RMSE = compareModelData(model,optimizer) 
```
