```python
def optimal_path_search(distance_matrix, start_state=None):
    """
    Performs brute force attempt to find the shortest possible path through a set of
    vectors. 
    distance_matrix: nxn symmetric matrix of distances. i, j is 
                     distance of ith point to jth
    start_state: State to start search with. Number in range [0, n]. If None will try all possible permutations.
    
    Returns: tuple of (min_permutation, distance for that permutation)
    """
    options = range(distance_matrix.shape[0])
    perms = itertools.permutations(options)
        
    min_perm = []
    min_dist = np.Inf
    
    max_perm = []
    max_dist = 0
    
    for perm in perms:
        if start_state is not None and perm[0] != start_state:
            continue
         
        print(perm)
        dist = get_path_length(perm, distance_matrix)
        if dist < min_dist:
            min_perm = perm
            min_dist = dist
            
        if dist > max_dist:
            max_perm = perm
            max_dist = dist
            
    return (min_perm, min_dist, max_perm, max_dist)

def get_path_length(perm, distance_matrix):
    dist = 0
    for i in range(len(perm)-1):
        dist = dist + distance_matrix[perm[i],perm[i+1]]
    return dist

def TSP_optimal_cos(matrix, numCities):
    tour = random.sample(range(numCities),numCities)
    distance = []
    besttour = tour
    bestdistance = get_path_length(tour, matrix)
    for temperature in numpy.logspace(0,3,num=200000)[::-1]:
        [i,j] = sorted(random.sample(range(numCities),2))
        newTour =  tour[:i] + tour[j:j+1] +  tour[i+1:j] + tour[i:i+1] + tour[j+1:]
        #oldDist = sum([math.sqrt(sum([(cities[newTour[(k+1) % numCities]][d] - cities[newTour[k % numCities]][d])**2 for d in [0,1] ])) for k in [j,j-1,i,i-1]])
        #newDist = sum([ math.sqrt(sum([(cities[tour[(k+1) % numCities]][d] - cities[tour[k % numCities]][d])**2 for d in [0,1] ])) for k in [j,j-1,i,i-1]])

        oldDist = get_path_length(tour, matrix)
        newDist = get_path_length(newTour, matrix)
        if math.exp( ( newDist - oldDist) / temperature) > random.random(): # cosine distance
            tour = copy.copy(newTour)
            distance.append(newDist)
            if newDist > bestdistance:
                besttour = newTour
                bestdistance = newDist
            
    return besttour, bestdistance, distance
```
